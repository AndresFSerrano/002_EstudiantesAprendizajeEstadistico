````python
class layer_nn():
    def __init__(self, act_func, nn_present, nn_before):
        self.theta = (2*np.random.random(( nn_present,nn_before))-1)
        self.B = 2*np.random.random((nn_present,1))-1
        self.act_func = act_func

    def output(self, Z, A, Ap):
        self.Z = Z
        self.A = A
        self.Ap = Ap

    def derivates(self, dtheta, db):
        self.dtheta = dtheta
        self.db = db

def func_activation(name, x):
    if name=="sigmoid":
        f = lambda x: 1/(1+np.exp(-x))
        fp = f(x)*(1-f(x))
        return f(x), fp


m, nx = np.shape(xtrain_)
A0 = xtrain_.T
# Parametros de entrada
topology = [nx, 10, 2, 23 , 20, 1]
func_name = ["sigmoid", "sigmoid", "sigmoid", "sigmoid", "sigmoid", "sigmoid"]
len(func_name)


# Inicializacion de la red
red = []
red.append(layer_nn(func_name[0],0, 0))
red[0].output(0,A0,0)

for i in range(1, len(topology)):
    red.append(layer_nn(func_name[i], topology[i], topology[i-1]))
````

## Pregunta2

2. Construir un generalizacion de la red, en el que entrada el valor inicial
y la red neuronal completa arroje la salida y la actualizacion de la red con los parametros deseados:

  ```
  A, nn = forward_pass(A0, nn_red)

 ```

````python
def forward_pass(A0, red):
  n=len(red)
  for i in range(1,n):
    Z = red[i].theta@A0 + red[i].B
    A,Ap = func_activation(red[i].act_func, Z )
    red[i].output(Z, A, Ap)
    A0 = A
  return A, red

A, nn=forward_pass(A0, red)
````

3. Encontrar la funcion de coste.


$$-\frac{1}{m} \sum\limits_{i = 1}^{m} (y^{(i)}\log\left(a^{[L] (i)}\right) + (1-y^{(i)})\log\left(1- a^{[L](i)}\right)) \tag{7}$$



````python
def costo(A0, Y):
    ji = Y*np.log(A0)+(1-Y)*np.log(1-A0)
    return ji.mean()
Ytrain = np.reshape(train_label,(1, 209))
Y = Ytrain
costo(red[-1].A, Y)
````

## Pregunta 4



4. Construir un codigo que permita realizar el BackwardPropagation

````python
def backward_pass(red,Y):
  m=Y.shape[1]
  L=len(red)-1

  # Para la primera capa de atrás hacia adelante
  A=red[L].A
  dAL=-(np.divide(Y,A)-np.divide(1-Y,1-A))
  dZ=dAL*red[L].Ap
  m_=1./m
  dtheta=m_ * dZ @ red[L-1].A.T
  db=m_*np.sum(dZ, axis=1, keepdims=True)
  red[L].derivates(dtheta, db)

  #Pa las otras capas
  for l in reversed(range(1,L)):
    dZ = (red[l+1].theta.T @ dZ) * red[l].Ap
    dtheta = m_ * dZ @ red[l-1].A.T
    db = m_ * np.sum(dZ, axis=1, keepdims=True)
    red[l].derivates(dtheta, db)
  return red

````

Aunque esto no se pide directamente, usando los datos de los gatos se puede entrenar a la red y hacer el backward propagation para verificar si el codigo, en efecto, funciona.

````python

A0 = xtrain_.T
Y = Ytrain

for repeticion in range(1000):
    A_out, red = forward_pass(A0, red)
    red = backward_pass(red, Y)
    alpha = 0.1

    for layer in red[1:]:
        layer.theta -= alpha * layer.dtheta
        layer.B -= alpha * layer.db

    if repeticion % 100 == 0:
        print(f"repeticion {repeticion}, costo: {-costo(red[-1].A, Y):.4f}")
````
