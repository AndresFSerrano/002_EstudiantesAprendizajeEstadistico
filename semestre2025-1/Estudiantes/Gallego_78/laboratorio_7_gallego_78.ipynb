- $\begin{equation}

\end{equation}$

- $\begin{equation}
db^{[l]} =  d{\cal Z}^{(i)[L]}  = dA^{(i)[l]} f'({\cal Z}^{(i)[l]} )
\end{equation}$
```
db_L =  m_*np.sum(dZ, axis=1, keepdims=True)
```
Los valores de dA pueden ser escritos como:
- $\begin{equation}
dA^{(i)[l-1]} = \Theta^{l} \cdot dZ^{(i)[l]}
\end{equation}$





$$ dZ^{[l]} = dA^{[l]} * f'^{[l]} (Z^{[l]}) $$
$$ d\Theta^{[l]} = \frac{\partial \mathcal{J} }{\partial \Theta^{[l]}} = \frac{1}{m} dZ^{[l]} A^{[l-1] T} \tag{1}$$

````python
class layer_nn():
    def __init__(self, act_fun, nlayer_present, nlayer_before):
        self.act_fun = act_fun
        self.n_present = nlayer_present
        self.n_before = nlayer_before
        
        if act_fun == 'relu':
            std = np.sqrt(2.0 / nlayer_before)
        else:
            std = np.sqrt(1.0 / nlayer_before)
        
        self.W = np.random.randn(nlayer_present, nlayer_before) * std
        self.b = np.zeros((nlayer_present, 1))
        self.dW = None
        self.db = None
        
    def forward(self, A_prev):
        self.A_prev = A_prev
        self.Z = np.dot(self.W, A_prev) + self.b
        self.A = act_function(self.Z, self.act_fun)
        return self.Z, self.A
    
    def backward(self, dA, learning_rate):
        m = dA.shape[1]
        dZ = dA * act_function_derivative(self.Z, self.act_fun)
        self.dW = (1/m) * np.dot(dZ, self.A_prev.T)
        self.db = (1/m) * np.sum(dZ, axis=1, keepdims=True)
        dA_prev = np.dot(self.W.T, dZ)
        self.W -= learning_rate * self.dW
        self.b -= learning_rate * self.db
        return dA_prev

class NeuralNetwork():
    def __init__(self, topology, activation_functions):
        self.topology = topology
        self.activation_functions = activation_functions
        self.L = len(topology) - 1
        
        if len(activation_functions) != len(topology):
            raise ValueError("El número de funciones de activación debe coincidir con el número de capas")
        
        self.layers = []
        for l in range(1, len(topology)):
            n_present = topology[l]
            n_before = topology[l-1]
            act_fun = activation_functions[l]
            layer = layer_nn(act_fun, n_present, n_before)
            self.layers.append(layer)
    
    def forward_pass(self, A0):
        A = A0
        self.cache = {'Z': [None] * (self.L + 1), 'A': [None] * (self.L + 1)}
        self.cache['A'][0] = A0
        
        for l in range(self.L):
            Z, A = self.layers[l].forward(A)
            self.cache['Z'][l+1] = Z
            self.cache['A'][l+1] = A
        
        return A, self
    
    def compute_cost(self, AL, Y):
        m = Y.shape[1]
        AL = np.clip(AL, 1e-15, 1 - 1e-15)
        cost = -(1/m) * np.sum(Y * np.log(AL) + (1 - Y) * np.log(1 - AL))
        cost = np.squeeze(cost)
        return cost
    
    def backward_pass(self, Y, learning_rate):
        m = Y.shape[1]
        AL = self.cache['A'][-1]
        dAL = -(np.divide(Y, AL) - np.divide(1 - Y, 1 - AL))
        dA = dAL
        for l in reversed(range(self.L)):
            dA = self.layers[l].backward(dA, learning_rate)
    
    def train(self, X, Y, learning_rate=0.01, epochs=1000):
        costs = []
        
        for i in range(epochs):
            AL, _ = self.forward_pass(X)
            cost = self.compute_cost(AL, Y)
            costs.append(cost)
            self.backward_pass(Y, learning_rate)
        
        return costs
    
    def predict(self, X):
        A, _ = self.forward_pass(X)
        if self.activation_functions[-1] == 'sigmoid':
            return (A > 0.5).astype(int)
        else:
            return A
    
    def accuracy(self, X, Y):
        predictions = self.predict(X)
        return np.mean(predictions == Y) * 100


def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

def linear(x):
    return x

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

def relu_derivative(x):
    return (x > 0).astype(float)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

def linear_derivative(x):
    return np.ones_like(x)

def act_function(x, activation):
    if activation is None or activation == 'linear':
        return linear(x)
    elif activation == 'sigmoid':
        return sigmoid(x)
    elif activation == 'relu':
        return relu(x)
    elif activation == 'tanh':
        return tanh(x)
    elif callable(activation):
        return activation(x)
    else:
        raise ValueError(f"Función de activación no reconocida: {activation}")

def act_function_derivative(x, activation):
    if activation is None or activation == 'linear':
        return linear_derivative(x)
    elif activation == 'sigmoid':
        return sigmoid_derivative(x)
    elif activation == 'relu':
        return relu_derivative(x)
    elif activation == 'tanh':
        return tanh_derivative(x)
    elif callable(activation):
        if hasattr(activation, 'derivative'):
            return activation.derivative(x)
        else:
            raise ValueError("Función personalizada debe tener método derivative")
    else:
        raise ValueError(f"Función de activación no reconocida: {activation}")

def visualize_sample_images(xtrain, train_label, num_samples=5):
    plt.figure(figsize=(15, 3))

    for i in range(num_samples):
        plt.subplot(1, num_samples, i + 1)
        img = xtrain[i]
        plt.imshow(img)
        plt.title(f"Label: {train_label[0, i]}\n{'Gato' if train_label[0, i] == 1 else 'No Gato'}")
        plt.axis('off')
    plt.tight_layout()
    plt.show()


data_train = "train_catvnoncat.h5"
data_test = "test_catvnoncat.h5"

train_dataset = h5py.File(data_train, "r")
test_dataset = h5py.File(data_test, "r")

xtrain_classes = np.array(train_dataset["list_classes"])
xtrain = np.array(train_dataset["train_set_x"])
train_label = np.array(train_dataset["train_set_y"])

test_classes = np.array(test_dataset["list_classes"])
xtest = np.array(test_dataset["test_set_x"])
test_label = np.array(test_dataset["test_set_y"])

# Preprocesamiento
# Redimensionar y normalizar
xtrain_ = np.reshape(xtrain, (xtrain.shape[0], -1)).T / 255.0
xtest_ = np.reshape(xtest, (xtest.shape[0], -1)).T / 255.0

train_label = train_label.reshape(1, -1)
test_label = test_label.reshape(1, -1)

# Visualizar algunas imágenes
print("\n Algunas imágenes de muestra")
visualize_sample_images(xtrain, train_label)

# Definir la arquitectura de la red
# 12288 entradas (64x64x3)
topology = [12288, 20, 7, 5, 1]  # Capas: [entrada, hidden1, hidden2, hidden3, salida]
activation_functions = [None, 'relu', 'relu', 'relu', 'sigmoid']

print(f"\nArquitectura de la red: {topology}")
print(f"Funciones de activación: {activation_functions}")

# Crear y entrenar la red neuronal
nn = NeuralNetwork(topology, activation_functions)

costs = nn.train(xtrain_, train_label, 
                learning_rate=0.0075, 
                epochs=2500)

train_accuracy = nn.accuracy(xtrain_, train_label)
test_accuracy = nn.accuracy(xtest_, test_label)

plt.figure(figsize=(10, 6))
plt.plot(costs)
plt.title('Costo durante el Entrenamiento')
plt.xlabel('Epoch')
plt.ylabel('Costo')
plt.grid()
plt.show()

print(f"Training Accuracy: {train_accuracy:.2f}%")
print(f"Test Accuracy: {test_accuracy:.2f}%")
print(f"Número de parámetros: {sum(layer.W.size + layer.b.size for layer in nn.layers):,}")
````
