15. ¿Existe una correlación significativa entre la longitud del pico y la masa corporal en las diferentes especies de pingüinos?


```python
correlation = penguins.groupby('species')[['body_mass_g', 'bill_length_mm']].corr().unstack().iloc[:, 1]
correlation = correlation.reset_index()
correlation.columns = ['Especie', 'Correlación']
print("Coeficientes de correlación de Pearson:")
print(correlation)
```

Sí hay cambios en la correlación por especie, para la especie Gentoo vemos que esta correlación es considerablemente mayor.

```python
# Crear grid de gráficos por especie
g = sns.lmplot(
    data=penguins,
    x='bill_length_mm',
    y='body_mass_g',
    hue='species',
    col='species',
    height=5,
    aspect=0.8,
    scatter_kws={'alpha': 0.6}
)

g.set_axis_labels('Longitud del pico (mm)', 'Masa corporal (g)')
plt.suptitle('Relación masa corporal - longitud del pico por especie', y=1.05)
plt.tight_layout()
plt.show()
```

17. Escoja una de las tendencias que haya observado en la cual haya una correlación entre las variables y ahora con un test de hipótesis determine si esta correlación puede ser explicada por aleatoriedad. Para esta pregunta puede usar las relaciones del libro "Probabilidad y Estadistica" de Marques de Cantu sección 12.6 donde hay un ejemplo 12.8 que les guia como usar las expresiones y hacer las conclusiones.

Vamos a utilizar la masa corporal y la longitud el pico.

H$_o$ : No existe correlación entre la masa corporal y la longitud del pico. \\
H$_1$ : Existe relación lineal entre la masa corporal y la longitud del pico.

```python
masa_corporal = penguins_clean['body_mass_g']
longitud_pico = penguins_clean['bill_length_mm']
corr, p_value = stats.pearsonr(masa_corporal, longitud_pico)
print("Coeficiente de correlación de Pearson:", corr)
print("Valor p:", p_value)
```

El valor p es extremadamente bajo. Con un nivel de significancia de $α$ = 0.05, tenemos evidencia para rechazar la hipótesis nula. Por tanto podemos afirmar estadísticamente que hay una relación lineal entre la masa corporal y la longitud del pico.
