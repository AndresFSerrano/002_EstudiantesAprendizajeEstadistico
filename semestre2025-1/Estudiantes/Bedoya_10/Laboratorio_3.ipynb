# Regresión lineal:

Supongamos que tenemos un sistema en el que existe un predicor con  **m** valores de entrenamiento,  asi:   $ (x^{(1)}, y^{(1)} ) , (x^{(2)}, y^{(2)}) ... (x^{(m)}, y^{(m)})$


|Training|Y      | X_1  |
|--------|----------|----------|
|1|$Y^{1}$ | $X_1^{1}$|
|2|$Y^{2}$ | $X_1^{2}$|
|.|.         | .        |
|.|.         | .        |
|.|.         | .        |
|m|$Y^{m}$ | $X_1^{m}$  |



Podemos definir un modelo lineal  como : $h(X) = \theta_0 + \theta_1 X$ con $(\theta_0 , \theta_1)$ parámetros. Nuestro objetivo es encontrar el conjunto de puntos  $(\theta_0 , \theta_1)$ que se encuentan más "cercano" a $Y$ para cada $X$.

Para la optimización, vamos a definir la función de coste **$J(\theta_1,\theta_2 )$** para las muestras de entrenamiento como aquella que garantiza la distancia euclidiana respecto a la hipotesis planteada, así:

\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( h_{\theta} (x^{(i)})-y^{(i)})^2
\end{equation}

Para encontrar los valores  $(\theta_0 , \theta_1)$ se necesita  minimizar la función de coste, que permite obtener los valores más cercanos,  esta minimización podrá ser realizada a través de diferentes métodos el más conocido es el gradiente descendente.

![](https://github.com/hernansalinas/Curso_aprendizaje_estadistico/blob/main/Sesiones/imagenes/fig00.png?raw=true=50x)




Supongamos un modelo lineal para realizar la predicción, así nuestro modelo estará basado en la siguiente hipotesis de trabajo:

$h(X) =\theta_0 + \theta_1 X$


Reemplando el modelo anterior en la función de coste, tenemos que:



\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( (\theta_0 + \theta_1 X)- y^{(i)})^2
\end{equation}

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

### 1) Se crea el dataframe

X_1 = [0,1,2,3,4,5]
Y = [0,1,2,3,4,5]
X_1 = pd.DataFrame(X_1)
Y = pd.DataFrame(Y)
df = pd.concat([Y, X_1], axis=1)
df.columns = ['Y', 'X_1']
df
````

````python
### 2) Se implementa la función de coste

def fun_coste(df,theta_0,theta_1):
  m = len(df)
  suma = 0
  for i in range(m):
    suma =+ ((theta_0 + theta_1*df.iloc[i,1])-df.iloc[i,0])**2
  return suma/(2*m)

fun_coste(df,0,2)
````

````python
### 3) Se evalua la función de coste para diferentes valores de theta1

theta_0 = 0
theta_1 = np.linspace(-10,10,1000)
J = []
for i in range(len(theta_1)):
  J.append(fun_coste(df,theta_0,theta_1[i]))

plt.plot(theta_1,J)
plt.title("Función de coste")
plt.xlabel("$\\theta_1$")
plt.ylabel("$J(\\theta_1$)")
plt.grid()
plt.show()
````

````python
### se halla el mínimo de la función de coste
theta_1 = theta_1[np.argmin(J)]
theta_1
````

# Análisis de la Función de Coste

#### Caso 1D: Se fija $\theta_0 = 0$ y se varía solo $\theta_1$

La hipótesis se reduce a:

$$
h_\theta(x) = \theta_1 x
$$

La función de coste es:

$$
J(\theta_1) = \frac{1}{2m} \sum_{i=1}^m \left( \theta_1 x_i - y_i \right)^2
$$

Como $x_i = y_i$:

$$
J(\theta_1) = \frac{1}{2m} \sum_{i=1}^m \left( \theta_1 x_i - x_i \right)^2 = \frac{1}{2m} \sum_{i=1}^m x_i^2 (\theta_1 - 1)^2
$$

Factorizando:

$$
J(\theta_1) = \frac{(\theta_1 - 1)^2}{2m} \sum_{i=1}^m x_i^2
$$

Para nuestros datos:

$$
\sum x_i^2 = 0^2 + 1^2 + 2^2 + 3^2 + 4^2 + 5^2 = 55, \quad m = 6
$$

Entonces:

$$
J(\theta_1) = \frac{55}{12} (\theta_1 - 1)^2
$$

El mínimo se alcanza en $\theta_1 = 1$.

---

#### Caso 2D: Se permiten variaciones en $\theta_0$ y $\theta_1$
Hipótesis completa:

$$
h_\theta(x) = \theta_0 + \theta_1 x
$$

La función de coste es:

$$
J(\theta_0, \theta_1) = \frac{1}{2m} \sum_{i=1}^m \left( \theta_0 + \theta_1 x_i - y_i \right)^2
$$

Como $y_i = x_i$:

$$
J(\theta_0, \theta_1) = \frac{1}{2m} \sum_{i=1}^m \left( \theta_0 + \theta_1 x_i - x_i \right)^2 = \frac{1}{2m} \sum_{i=1}^m \left( \theta_0 + x_i (\theta_1 - 1) \right)^2
$$

# Gradiente Descendente


Para determinar el mínimo,  a través del gradiente descendente puede ser aplicada el siguiente algoritmo:

- Proponer un número aleatorios inicial $\omega_i$
- Para descender al mínimo de la función, encontremos un  valor para el cual
el deriva de la función descenciende, asi:
\begin{equation}
\omega_{i+1} = \omega_{i} - \alpha \frac{\mathrm{d}f(\omega_i)}{\mathrm{d}\omega}
\end{equation}

donde, $\alpha$ es conocido como la tasa de aprendizaje del algoritmo.

- Evaluar $f(\omega_{i+1})$

- Iterar hasta encontrar el mínimo de la función

````python
### 6) Se implementa el gradiente descendiente para una función de un parámetro
x_0 = np.random.uniform(0,5)
def f(x):
  return (x-4)**2

def df(x):
  return 2*(x-4)

def grad_desc(x_0, alpha, df, epsilon):
  iter = 1
  while True:
    x = x_0 - alpha*df(x_0)
    if abs(x-x_0) < epsilon:
      break
    x_0 = x
    iter = iter +1
  return x,iter

print(f"Con alpha = 0.1 y x_0 = {x_0}, el mínimo de la función es {grad_desc(x_0,0.1,df,1E-4)[0]} con {grad_desc(0,0.1,df,1E-4)[1]} iteraciones")
print(f"Con alpha = 0.5 y x_0 = {x_0}, el mínimo de la función es {grad_desc(x_0,0.5,df,1E-4)[0]} con {grad_desc(0,0.5,df,1E-4)[1]} iteraciones")
print(f"Con alpha = 0.9 y x_0 = {x_0}, el mínimo de la función es {grad_desc(x_0,0.9,df,1E-4)[0]} con {grad_desc(0,0.9,df,1E-4)[1]} iteraciones")
````

# Modelo de *machine learning*: Solución general

Un modelo general para solucionar un problema de machinine learning puede ser estructurado como sigue:


a. Eliger el modelo a emplear:
\begin{equation}
h(X,\Theta)
\end{equation}

- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:

    \begin{equation}
    h(X) = (\theta_0 + \theta_1 X)
    \end{equation}


b. Eligir la funcion de coste :
-  Metrica Ecuclidiana:
\begin{equation}
J(\Theta)=\frac{1}{2m} \sum_{i=1}^m ( h_{\theta} (X^{(i)})-y^{(i)})^2
\end{equation}
- [Lista de funciones de coste que pueden ser empleadas](https://jmlb.github.io/flashcards/2018/04/21/list_cost_functions_fo_neuralnets/)


c. Aplicar el gradiente descendente iterativamente, hasta encontrar el minimo:  

\begin{equation}
\Delta \vec{\Theta} =  - \alpha \nabla J(\Theta)
\end{equation}



- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:


\begin{equation}
\theta_0 := \theta_0 - \alpha \frac{\partial J}{\partial \theta_0}
\end{equation}

\begin{equation}
\theta_1 := \theta_1 - \alpha \frac{\partial J}{\partial \theta_1}
\end{equation}

````python
### 8) Se grafican los datos
X = np.linspace(0, 1, 100)
y = 0.2 + 0.2*X + 0.02*np.random.random(100)

plt.plot(X,y,'.')
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
````

````python
### Se crea el dataframe con los datos proporcionados
df = pd.DataFrame({'x': X, 'y': y})
df
````

````python
### Se define la función de coste cuadrático medio y se grafica
def fun_coste2(dataframe, theta0, theta1):
    x = dataframe['x'].values
    y = dataframe['y'].values
    pred = theta0 + theta1 * x
    return np.mean((pred - y)**2)

theta_0_vals = np.linspace(-15, 15, 100)
theta_1_vals = np.linspace(-15, 15, 100)
T0, T1 = np.meshgrid(theta_0_vals, theta_1_vals)

J = fun_coste(df,T0,T1)
fig = plt.figure()
ax = plt.axes(111,projection='3d')
ax.plot_surface(T0, T1, J )
ax.contour(T0, T1, J, 200,   linestyles="solid")
ax.set_xlabel("$\\theta_0$")
ax.set_ylabel("$\\theta_1$")
plt.show()
````

````python
### Se minimiza por el método del gradiente descendiente
Theta = np.random.rand(2) * 4 - 2  # Inicio aleatorio en [-2, 2]
h = 1e-5
lr = 0.1
n_iter = 1000

for _ in range(n_iter):
    grad = np.zeros(2)

    for i in range(2):
        _T = Theta.copy()
        _T[i] += h
        deriv = (fun_coste2(df, _T[0], _T[1]) - fun_coste2(df, Theta[0], Theta[1])) / h
        grad[i] = deriv

    Theta -= lr * grad

theta0_opt, theta1_opt = Theta
print(f"Valores óptimos:")
print(f"theta_0 = {theta0_opt:.5f}")
print(f"theta_1 = {theta1_opt:.5f}")
print(f"Error cuadrático medio: {fun_coste2(df, theta0_opt, theta1_opt):.5e}")

plt.figure(figsize=(8, 5))
plt.plot(df['x'], df['y'], '.', label="Datos", alpha=0.7)
y_pred = theta0_opt + theta1_opt * df['x']
plt.plot(df['x'], y_pred, 'r-', label="Regresión lineal", linewidth=2)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Ajuste Lineal")
plt.legend()
plt.grid(True)
plt.show()
````

````python
### Se compara con la solución de sklearn
from sklearn.linear_model import LinearRegression

model = LinearRegression()
X_reshape = df['x'].values.reshape(-1, 1)
y_vals = df['y'].values

model.fit(X_reshape, y_vals)

theta0_sklearn = model.intercept_
theta1_sklearn = model.coef_[0]

print("\n[sklearn] Resultados con LinearRegression:")
print(f"theta_0 = {theta0_sklearn:.5f}")
print(f"theta_1 = {theta1_sklearn:.5f}")

def mse(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

y_pred_sklearn = model.predict(X_reshape)
error_sklearn = mse(y_vals, y_pred_sklearn)
print(f"Error cuadrático medio (sklearn): {error_sklearn:.5e}")

plt.figure(figsize=(8, 5))
plt.plot(df['x'], df['y'], '.', label="Datos", alpha=0.7)
plt.plot(df['x'], y_pred, 'r-', label="Gradiente descendente", linewidth=2)
plt.plot(df['x'], y_pred_sklearn, 'g--', label="LinearRegression (sklearn)", linewidth=2)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gradiente descendente vs sklearn")
plt.legend()
plt.grid(True)
plt.show()
````
